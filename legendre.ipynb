```python
from scipy.special import legendre
import matplotlib.pyplot as plt
import numpy as np
L = 2
generator = legendre(L)
th = np.linspace(0,180,1801)
x = np.cos(np.deg2rad(th))
y = generator(x)**2 #P_L^2(cos(theta))
plt.plot(th,y,label="d$\sigma$/d$\\theta$")
plt.plot(th,y*np.sin(np.arccos(x)),label="d$\sigma$/d$\Omega$")
plt.legend()
plt.xlabel("COM angle")
plt.ylabel("Yield")
plt.show()
```
